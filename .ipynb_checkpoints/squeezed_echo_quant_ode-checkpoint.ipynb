g(t) is defined as:

\begin{equation}
g(t) = \mu_1 \delta (t) + \mu_2 \delta (t-\Delta)
\end{equation} 



In the paper, $\mu_1 = 5$, $\mu_2 = 2.5$ and $\Delta = 10$.

The initial state of the oscillator is: $Q_0 = 25$ and $P_0 = 100$


we are given that: 

\begin{equation}
\Xi(t) = \langle \Psi  |Q^2|\Psi  \rangle 
\end{equation}


\begin{equation}
\dot{\Xi}(t) = -i\hbar + 2 \langle \Psi  |QP|\Psi  \rangle 
\end{equation}


\begin{equation}
\ddot{\Xi}(t) = 2 \langle \Psi  |P^2|\Psi  \rangle - 2[\Omega^2 - 2 g(t)] \Xi
\end{equation} 

and finally,  

\begin{equation}
\dddot{\Xi} + 4\dot{\Xi}[\Omega^2 - 2g(t)] - 4\dot{g}(t)\Xi = 0 
\end{equation} 

With the above equations and parameters given, we can calculate the initial conditions for solving ODE: $\Xi(t=0)$, $\dot{\Xi}(t=0)$, and $\ddot{\Xi}(t=0)$.

```python
def gt_func(mu1, mu2, t_, Delta_):
    #t_ls = np.linspace(0,3*Delta_, 301)
    #gt_func_val = mu_1*signal.unit_impulse(len(t_ls)) + mu_2*signal.unit_impulse(len(t_ls),np.where(t_ls==Delta))
    if t_ == 0:
        gt_val = mu_1#*signal.unit_impulse(len(t_ls))
    elif t_ == Delta_:
        gt_val = mu_2#*signal.unit_impulse(len(t_ls),np.where(t_ls==Delta))
    else:
        gt_val = 0
    return gt_val
    
def dot_gt_func(mu1, mu2, t_, Delta_):
    #t_ls = np.linspace(0,3*Delta_, 301)
    #gt_func_val = mu_1*signal.unit_impulse(len(t_ls)) + mu_2*signal.unit_impulse(len(t_ls),np.where(t_ls==Delta))
    if t_ == 0:
        gt_val = -mu_1#*signal.unit_impulse(len(t_ls))
    elif t_ == Delta_:
        gt_val = -mu_2#*signal.unit_impulse(len(t_ls),np.where(t_ls==Delta))
    else:
        gt_val = 0
    return gt_val

def dSdt(t, S):
    f1, f2, f3 = S
    return [f2,
           f3,
           -4 * ((omega**2)- 2 * gt_func(mu_1, mu_2, t, Delta))*f2 + 4 * dot_gt_func(mu_1, mu_2, t, Delta) *f1]

f1_0 = Xi_init_val
f2_0 = dot_Xi_init_val
f3_0 = ddot_Xi_init_val

S_0 = (f1_0, f2_0, f3_0)
```

```python
omega_ls = [5,7.5,10]
t = np.linspace(0, 3*Delta, 301)
fig, ax = plt.subplots(len(omega_ls),1,figsize=(5,len(omega_ls)*6),
                       tight_layout = True)
for i in range(len(omega_ls)):
    omega = omega_ls[i]
    sol = odeint(dSdt, y0=S_0, t=t, tfirst=True)
    ax[i].plot(t/Delta,sol.T[0])

```

```python
def lorentzian_dist(x, x0, a, gam):
    return a * gam**2 / ((0.5 * gam)**2 + (x - x0)**2)


### my code that works ###
def random_choice(x_ls, distributions, N0):
    random_number = random.choices(x_ls, distributions, k=int(N0))
    return random_number

omega_0 = 7.5
gamma = 0.5
N0 = 1e2
dist_x_ls = np.r_[np.linspace(omega_0-1000, omega_0 - 5, 100), 
                  np.linspace(omega_0-5, omega_0 + 5, 1000),
                  np.linspace(omega_0+5, omega_0 + 1000, 100)]

dist = lorentzian_dist(dist_x_ls, omega_0,
                       1/np.pi, gamma)

#print (dist)
rand_sample = random_choice(dist_x_ls, dist, N0)
print (np.shape(rand_sample))

```

```python
t = np.linspace(0, 3*Delta, 301)

sum_sol = 0
for i in range(len(rand_sample)):
    omega = rand_sample[i]
    sol = odeint(dSdt, y0=S_0, t=t, tfirst=True)
    sum_sol += sol.T[0]
    #ax[i].plot(t/Delta,)
```
