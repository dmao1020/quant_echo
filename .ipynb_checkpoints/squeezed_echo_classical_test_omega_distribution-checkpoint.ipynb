```python
import scipy
from scipy.constants import pi
import numpy as np
import math
import time
import random
import os
#import seaborn as sns
import matplotlib.pylab as plt
```

```python
# classical oscillators
# defining U(t) function where 0 < t < delta
def U_func(omega, t, p_0, q_0, mu_1):
    return q_0*np.cos(omega*t) + np.sin(omega*t)*((p_0 + 2*mu_1*q_0)/omega)

def ut_sum_func(sample_ls, t_ut, p0, q0, mu1):
    Ut_sum = 0.0
    for i in range(len(sample_ls)):
        omega_i = sample_ls[i]
        Ut_i = U_func(omega_i, t_ut, p0, q0, mu1)
        Ut_sum += Ut_i

    return Ut_sum


#Defining Q(t) function where t > delta
def Q_func(omega, t, p_0, q_0, mu_1, mu_2, delta_):
    U_t_result = U_func(omega, t, p_0, q_0, mu_1)
    U_delta_result = U_func(omega, delta_, p_0, q_0, mu_1)
    return U_t_result + (np.sin(omega*(t-delta_))*((2*mu_2*U_delta_result)/omega))
    

def qt_sum_func(sample_ls, t_qt, p0, q0, mu1, mu2, Delta):
    Qt_sum = 0.0
    for i in range(len(sample_ls)):
        omega_i = sample_ls[i]
        
        Qt_i = Q_func(omega_i, t_qt, p0, q0, mu1, mu2, Delta)
        Qt_sum += Qt_i
    return Qt_sum



# momentum functions 

def dot_ut(omega, t, p_0, q_0, mu_1):
    return (-omega*q_0*np.sin(omega*t))+(p_0+2*mu_1*q_0)*np.cos(omega*t)

def dot_qt(omega, t, p_0, q_0, mu_1, mu_2, delta_):
    Ut_dot= dot_ut(omega, t, p_0, q_0, mu_1)
    U_delta_ = U_func(omega, delta_, p_0, q_0, mu_1)
    return Ut_dot + 2*mu_2*U_delta_*np.cos(omega*(t-delta_))

# define pdf
    
def lorentzian_dist(x, x0, a, gam):
    return a * gam**2 / ((0.5 * gam)**2 + (x - x0)**2)


def gaussian_dist(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma)**2) / (np.sqrt(2 * np.pi) * sigma)

```

```python
### my code that works ###
def random_choice(x_ls, distributions, N0):
    random_number = random.choices(x_ls, distributions, k=int(N0))
    return random_number
```

```python
### inverse transform sampling ###
def inverse_transform_sampling(num_samples, x0, a, gam):
    u = np.random.uniform(size=int(num_samples))
    samples = x0 + a * gam * np.tan(np.pi * (u - 0.5))
    return samples
```

```python
##### sampling lorentzian with gaussian distribution #####
def sample_lorentzian_gauss(num_samples, x0, a, gam, gaussian_sig_scale):
    samples = []
    while len(samples) < num_samples:
        # Generate samples from the proposal Gaussian distribution with a larger standard deviation
        x_proposal = np.random.normal(loc=x0, scale=gam * gaussian_sig_scale)  # Increase the scaling factor (e.g., 10)

        # Calculate the acceptance ratio
        acceptance_ratio = lorentzian_dist(x_proposal, x0, a, gam) / gaussian_dist(x_proposal, x0, gam*gaussian_sig_scale)  # Update the scaling factor
        #acceptance_ratio = lorentzian_dist(x_proposal, x0, a, gam) / gaussian_dist(x_proposal, x0, gam)

        # Generate a uniform random number
        u = np.random.uniform()

        # Accept or reject the sample
        if u < acceptance_ratio:
            samples.append(x_proposal)

    return samples
```

```python
##### sampling lorentzian with uniform distribution #####
def sample_lorentzian_uniform(x0, a, gam, x_min, x_max, num_samples):
    
#     # Define the range of x values to sample from
#     x_min = -10
#     x_max = 10
#     num_samples = 1000

    # Generate uniform random numbers between 0 and 1
    u = np.random.uniform(size=int(num_samples))

    # Apply inverse transform sampling
    x = x_min + (x_max - x_min) * u
    sample = lorentzian_dist(x, x0, a, gam)
    
    return sample
```

```python
### parameters for the Lorentzian distribution

omega_0 = 7.5
gamma = 0.5
a = 1/np.pi
gauss_sig_scale = 1
### parameters of the forces acting on the oscillators
mu1 = 5
mu2 = 2.5
Delta = 10

### initial condition of the oscillator
p0 = 10
q0 = 25
N0 = 1e4
```

For $0 < t < \Delta$ Q(t) = U(t) and U(t) is defined as:

\begin{equation}
U(t) = Q_0 \cos \Omega t + \frac{(P_0 + 2 \mu_1 Q_0)}{\Omega} \sin \Omega t
\end{equation}

, and for $t > \Delta$
\begin{equation}
Q(t) = U(t) + \frac{2\mu_2 U(\Delta)}{\Omega} \sin \Omega (t-\Delta)
\end{equation}


The $\overline{Q(t)}$ as a function of $\t/\Delta$ is plotted, where we are taking the average of an ensemble of $N_0 = 10^4$ classical oscillators. 

The frequency ($\Omega$) of each classical oscillator is randomly sampled from  the Lorentzian distribution of frequencies. The Lorentzian distribution is defined as $\frac{\gamma/ \pi}{(\Omega - \Omega_0)^2 + (\gamma / 2)^2}$. 

The scatter graphs are showing momentum ($P$) as a function of $\Omega_0 Q$, where $P = \delta Q/ \delta t$. 

For $0 < t < \Delta$:
\begin{equation}
\delta U/ \delta t = \dot{U}(t) = -\Omega Q_0 \sin \Omega t +(P_0 + 2 \mu_1 Q_0)\cos \Omega t
\end{equation}

, and for $t > \Delta$
\begin{equation}
\dot{Q}(t) = \dot{U}(t)  + 2\mu_2 U(\Delta) \cos \Omega (t-\Delta)
\end{equation}


For all of the plots, these are the settings for all the parameters:

For Lorentzian distribution:
$\Omega_0 = 7.5$, and $\gamma = 0.5$

For the oscillators: There are $N0 = 10^4$ oscillator; each oscillator has the initial conditions of $Q_0 = 25$ and $P_0 = 10$

For the parametric force acting on the oscillators:
$\Delta = 10$, $\mu_1 = 5$ and  $\mu_1 = 2.5$


```python
### sample from lorentzian distribution ###
"""
def lorentzian_dist(x, x0, a, gam):
    return a * gam**2 / ((0.5 * gam)**2 + (x - x0)**2)
    

def random_choice(x_ls, distributions, N0):
    random_number = random.choices(x_ls, distributions, k=int(N0))
    return random_number

"""

# define a list of omega values
dist_x_ls = np.r_[np.linspace(omega_0-1000, omega_0 - 5, 100), 
                  np.linspace(omega_0-5, omega_0 + 5, 1000),
                  np.linspace(omega_0+5, omega_0 + 1000, 100)]

# use the list of omega values ^ 
# and the defined lorentzian distribution function (lorentzian_dist(x, x0, a, gam))
# to calculate a the probability density function
dist = lorentzian_dist(dist_x_ls, omega_0,
                       1/np.pi, gamma)

# random sampling
rand_sample = random_choice(dist_x_ls, dist, N0)



print (np.min(rand_sample), np.max(rand_sample))

##### plot randomly sampled omega #####
fig, ax = plt.subplots(1,2,figsize=(5*3, 6),
                       gridspec_kw={'width_ratios': [2, 1]},
                       tight_layout = True)

ax[0].plot(dist_x_ls, dist, label = "Lorentzian Function")
#plt.xlim(0,15)
#### start sampling for omega #####
num_bins = 500
red_samples = [sample_i for sample_i in rand_sample if abs(sample_i) < 500]

n, bins, patches = ax[0].hist(red_samples, num_bins, 
                            density = 1,
                            color = "orange", 
                            alpha = 0.7, label = r"Histogram of randomly sampled $\Omega$ ($N_0 = $ %s)"%(int(N0)))
print (np.min(rand_sample), np.max(rand_sample))
ax[0].set_xlim(0,15)
ax[0].set_ylabel("Frequency", fontsize=15)
ax[0].set_xlabel(r"Sampled $\Omega$", fontsize=15)
ax[0].legend()


#####
t_mesh = 1000
t_ut = np.linspace(0, Delta, t_mesh)
t_qt = np.linspace(Delta, 3*Delta, 2*t_mesh)


Ut_sum = ut_sum_func(rand_sample, t_ut, p0, q0, mu1)
Qt_sum = qt_sum_func(rand_sample, t_qt, p0, q0, mu1, mu2, Delta)

ax[1].plot(t_ut/Delta, Ut_sum/len(rand_sample))
ax[1].plot(t_qt/Delta, Qt_sum/len(rand_sample))

ax[1].set_ylabel(r"$\overline{Q}$", rotation=0, fontsize=15)
ax[1].set_xlabel(r"$t/\Delta$", fontsize=15)



# time snap shots
t_snapshot_ls = [0.5*Delta, 0.8*Delta, Delta, 1.5*Delta, 2*Delta, 2.5*Delta]

fig, ax = plt.subplots(2,int(len(t_snapshot_ls)/2),
                       figsize=(len(t_snapshot_ls)*5/2, 10), 
                       tight_layout = True)
ax_coord = [[0,0],[0,1],[0,2],[1,0],[1,1],[1,2]]

for j in range(len(t_snapshot_ls)):
    ax_x, ax_y = ax_coord[j]
    t_i = t_snapshot_ls[j]
    if t_i < Delta:
        qp_ls = np.zeros((len(rand_sample),2))
        for i in range(len(rand_sample)):
            omega_i = rand_sample[i]
            Ut_i = U_func(omega_i, t_i, p0, q0, mu1)
            dot_ut_i = dot_ut(omega_i, t_i, p0, q0, mu1)
            
            qp_ls[i,:] =[omega_0*Ut_i, dot_ut_i]
        ax[ax_x, ax_y].scatter(qp_ls[:,0],qp_ls[:,1], s= 2)
        
    elif t_i >= Delta:
        qp_ls = np.zeros((len(rand_sample),2))
        for i in range(len(rand_sample)):
            omega_i = rand_sample[i]
            Qt_i = Q_func(omega_i, t_i, p0, q0, mu1, mu2, Delta)
            dot_qt_i = dot_qt(omega_i, t_i, p0, q0, mu1, mu2, Delta)
            
            qp_ls[i,:] =[omega_0*Qt_i, dot_qt_i]
        
        ax[ax_x, ax_y].scatter(qp_ls[:,0],qp_ls[:,1], s= 2)
    
    ax[ax_x, ax_y].set_title(r"$t = %s \Delta$"%(t_i/Delta), fontsize=15)
    ax[ax_x, ax_y].set_ylabel(r"$P$", fontsize=15)
    ax[ax_x, ax_y].set_xlabel(r"$\Omega_0 Q$", fontsize=15)
    
    ax[ax_x, ax_y].set_xlim(-500,500)
    ax[ax_x, ax_y].set_ylim(-500,500)
```
